## Subplots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(0, 10)

x1 = np.cos(n)
x2 = np.sin(n)
print(x1)
print(x2)
```

```python
plt.subplot(1, 2, 1)
plt.stem(n, x1)
plt.title("Cos")

plt.subplot(1, 2, 2)
plt.stem(n, x2)
plt.title("Sin")

plt.suptitle("Sinusoidal Functions")
plt.show()
```
