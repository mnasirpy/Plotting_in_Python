## Markers

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.randint(2, 8, size=(7))
```

```python
print(y)
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(y, marker='*', color='g')
plt.show()
```

```python
plt.plot(y, marker='D', color='r')
plt.show()
```

```python
plt.plot(y, marker='+', color='y', linestyle=':')
plt.show()
```

```python
plt.plot(y, marker='s', color='b', linestyle='-.')
```
