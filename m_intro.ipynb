```python
import matplotlib
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 6])
y = np.array([0, 250])
```

```python
print(x)
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
plt.plot(x, y, 'or')
plt.show()
```

```python
plt.stem(x, y, 'o')
plt.show()
```

```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])

plt.plot(x, y, 'g+')
plt.show() 
```

```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])

plt.plot(x, y, 'r--')
plt.show() 
```

```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])

plt.stem(x, y, 'g')
plt.show() 
```
